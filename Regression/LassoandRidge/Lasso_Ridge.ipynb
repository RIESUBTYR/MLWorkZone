```python
import pandas as pd
import seaborn as sns
import numpy as nps
import matplotlib.pyplot as plt
from sklearn import *

#Opening dataset

data = pd.read_csv("train.csv")

print(data.shape)

print(data.columns)

data.head()
```

```python
#Creating Training and Testing datasets

Data_x = data[['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']]

Data_y=House_data.loc[:,'price_range']

#split train - test data
train_x,test_x,train_y,test_y=model_selection.train_test_split(Data_x,Data_y,train_size=0.8,random_state=1)
```

```python
#train model
regr=linear_model.LinearRegression()
regr.fit(train_x,train_y)


pred_y=regr.predict(test_x)

# pred_y=[round(num) for num in pred_y]

# pred_y=[3 if element>=4 else element for element in pred_y ]
# pred_y=[0 if element<=0 else element for element in pred_y ]


#mse r2score
mse=metrics.mean_squared_error(test_y,pred_y)
rscore=metrics.r2_score(test_y,pred_y)
print('Mean Square Error: {}\n Correlation score: {}'.format(mse,rscore) )
```

```python

# graphs
plt.scatter(test_x['battery_power'],test_y,color='orange')
plt.scatter(test_x['battery_power'],pred_y,color='blue')
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#Lasso Regression

lasso=linear_model.Lasso(alpha=1)
lasso.fit(train_x,train_y)


pred_y=lasso.predict(test_x)

# pred_y=[round(num) for num in pred_y]
# pred_y=[3 if element>=4 else element for element in pred_y ]
# pred_y=[0 if element<=0 else element for element in pred_y ]

#mse r2score
mse=metrics.mean_squared_error(test_y,pred_y)
rscore=metrics.r2_score(test_y,pred_y)
print('Mean Square Error: {}\n Correlation score: {}'.format(mse,rscore) )

```

```python
# graphs
plt.scatter(test_x['battery_power'],test_y,color='orange')
plt.scatter(test_x['battery_power'],pred_y,color='blue')
plt.show()

```

```python
#Ridge Regression

ridge=linear_model.Ridge(alpha=1)
ridge.fit(train_x,train_y)


pred_y=ridge.predict(test_x)

# pred_y=[round(num) for num in pred_y]

# pred_y=[3 if element>=4 else element for element in pred_y ]
# pred_y=[0 if element<=0 else element for element in pred_y ]

#mse r2score
mse=metrics.mean_squared_error(test_y,pred_y)
rscore=metrics.r2_score(test_y,pred_y)
print('Mean Square Error: {}\n Correlation score: {}'.format(mse,rscore) )
```

```python

#graphs
plt.scatter(test_x['battery_power'],test_y,color='orange')
plt.scatter(test_x['battery_power'],pred_y,color='blue')
plt.show()
```
